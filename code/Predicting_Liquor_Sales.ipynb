
# Predicting Liquor Sales 

### Executive Summary

For this project I simulate an Iowa Liquor Store looking to expand their operations by starting a new store in Iowa. To help identify the best location for a new store, I analyzed transaction data provided by the Iowa Department of Commerce, and derived inference analysis to better understand the relationship in the data collected. Based on this data, I will develop a linear model to project total sales on a yearly basis.

We will successfully identify a county to target our expansion, by comparing the results of the model to the county who in 2015 had the most amount of total sales: Polk county.



#### Assumptions made on the data:

•	Only the quantitative variables that could potentially have correlation helpful to solve the problem, and among those are some with high enough relation to each other that would hurt the accuracy of the model. For that reason, only one of the three variables that describe volume was chosen: volume per bottle (ml).
•	Bottle cost and bottle retail share a consistent difference of 33% across the board, indicating a perfect relation. Once again, only one was chose: Bottle Retail price. An additional assumption with this category is a flat profit margin for all products on the store’s side.
•	Duplicates on the data were dropped from the analysis, but considering the same transaction on the same day would be indistinguishable in this dataset, it bears mentioning.



#### Methodology:

•	The data was downloaded from Iowa’s Dept. of Commerce, and underwent a basic evaluation where several inaccuracies in the data were identified and corrected (or removed from the model). For consistency in units, only data from 2015 is considered.
•	Polk county was selected to be removed as the control variable to compare other counties to because it displayed the most total sales in 2015 in the data set.
•	The variables selected by the task manager (Price per Bottle and Total Bottles Sold) were isolated, along with volume per bottle (ml).
•	A linear model was developed, and the data was split into training and testing sets. The variables selected were analyzed on a county-as-location basis and a list of coefficients (in relation to Polk county) were computed for each county.



#### Conclusions:

According to the model created, the best performers when compared to Polk county are:
- Dallas county, with a coefficient of 54.9
- Carroll county, with a coefficient of 10.4
- Sioux county, with a coefficient of 8.3
 
The coefficients are the predicted amount of extra dollars per purchase order to the state of Iowa, when compared to the purchase order from Polk county.

For that reason, the model recommends Dallas, Carroll, Sioux, and Johnson counties to start the next store.
